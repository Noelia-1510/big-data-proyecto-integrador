4. Realizar las siguientes consultas sobre una sola tabla. Los nombres de las columnas deben mostrarse en español:
    <b>A.</b>  Listar todos los registros de la tabla, visualizando las columnas type, title_content, country, rating, duration, listed_in.</br>
<font color='BlueViolet'>
SELECT type AS "tipo", title_content AS "título_del_contenido", country AS "país",
rating AS "clasificación", duration AS "duración", listed_in AS "categoría" 
FROM content;
</font></br>
    <b>B.</b>  Mostrar todas las filas de la tabla, presentando las columnas id_content, title_production, genre, runtime, imdb_Score.</br>
<font color='BlueViolet'>
SELECT id_content AS "id_del_contenido", title_production AS "título_de_la_producción",
genre AS "género", runtime AS "tiempo_de_ejecución", imdb_score AS "puntuación_en_imdb"
FROM production;
</font></br>
    <b>C.</b>  Generar un listado de todas las películas en las que participe Argentina, mostrando las columnas id_content, type, title_content, director, cast, rating, duration, listed_in.</br>
<font color='BlueViolet'>
SELECT id_content AS "id_del_contenido", type AS "tipo", title_content AS "título_del_contenido", director AS "director", cast AS "elenco", rating AS "clasificación", duration AS "duración", listed_in AS "categoría"
FROM content
WHERE country LIKE "%Argentina%";
</font></br>
    <b>D.</b>  Presentar un listado con las columnas id_production, title_production, genre, language, imdb_score de los registros de la tabla en las que el imdb_score sea superior a 7.5.</br>
<font color='BlueViolet'>
SELECT id_production AS "id_de_la_producción", title_production AS "título_de_la_producción",
genre AS "género", language AS "lenguaje", imdb_score AS "puntuación_en_imdb"
FROM production
WHERE imdb_score > 7.5
ORDER BY imdb_score DESC;
</font></br>
    <b>E.</b>  Mostrar las columnas id_content, type, title_content, listed_in de los registros en los que el rating público general o admitido para todas las edades.</br>
<font color='BlueViolet'>
SELECT id_content AS "id_del_contenido", type AS "tipo", title_content AS "título_del_contenido", listed_in AS "categoría"
FROM content
WHERE rating IN ("G", "TV-G");
</font></br>
    <b>F.</b>  Obtener una lista con las columnas id_production, title_production, genre de los registros en que duration sea menor o igual a 105 minutos pero que además tenga una duración de al menos 60 minutos.</br>
<font color='BlueViolet'>
SELECT id_production AS "id_de_la_produción", title_production AS "título_de_la_produción", genre AS "género"
FROM production
WHERE runtime BETWEEN 60 AND 105
ORDER BY runtime;
</font></br>
    <b>G.</b>  Obtener la cantidad de registros discriminando por type y rating, ordenados alfabéticamente por rating y por type.</br>
<font color='BlueViolet'>
SELECT type AS "tipo", rating AS "clasificación",
COUNT(id_content) AS "cantidad"
FROM content
WHERE rating IS NOT NULL
GROUP BY type, rating
ORDER BY type, rating;
</font></br>
    <b>H.</b>  Mostrar la cantidad  de películas en los idiomas portugués, francés, alemán, italiano y español, ordenados por cantidad, de mayor a menor.</br>
<font color='BlueViolet'>
SELECT language AS "idiomas",
COUNT(id_content) AS "cantidad"
FROM production
WHERE language IN ("Portuguese", "French", "German", "Italian", "Spanish")
GROUP BY language
ORDER BY COUNT(id_content) DESC;
</font></br>
