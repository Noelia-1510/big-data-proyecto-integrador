    <b>i.</b> Debe crear un **primer método** para la conexión `get_connetion()` y que retorne la conexión (variable que llamaremos `conn`) y el cursor (variable que llamaremos `cur`).
    <b>ii.</b> Debe crear un **segundo método** para la desconexión `close_connetion()`. Debe cerrar ambas variables creadas anteriormente de forma adecuada.

```
# Des/conexión

import sqlite3 as sq3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import display  # para mostrar display con el estilo DF de Jupiter Notebook


# CONEXION
def get_connection():
    global db_netflix
    global conn
    global cur
    try:
        db_netflix = '/work/Mi_proyecto/data/sql_lite/netflix_oscar.db'
        conn = sq3.connect(db_netflix)
        cur = conn.cursor()
        print('La conexión a la base de datos ha sido exitosa!')
        return conn, cur
    except (Exception, sq3.Error) as error:
        print(f'Error al conectarse a la base de datos: , {err}')

# DESCONEXION
def close_connection():
    cur.close()
    conn.close()
    print('Base de datos desconectada exitosamente')

get_connection()
print()
close_connection()
```

```
# Matplotlib
# Para ejecutar esta celda debe haberse ejecutado la tercera celda llamada # Consultas en DF

get_connection()
df_content = pd.read_sql('SELECT * FROM content', conn)

custom_colors = ["#663399","#FF69B4"] # Crea una paleta de colores personalizada en formato hexadecimal
fig, ax = plt.subplots()
df_content.groupby(['type'])['type'].count().plot(kind='pie', autopct='%1.2f%%', colors=custom_colors, startangle=90)
ax.set_title('Relación entre series y películas')
plt.show()

close_connection()
```

<b>ii.</b> Realizar un gráfico a partir de los países que tienen mayor cantidad total de títulos en la plataforma, desglosando por tipo de contenido, mostrando la información sólo para los primeros 10 países (mostrar el desglose por tipo de contenido), Obvie las cooperaciones entre países. <font color='blue'>**Tip puede usar el método `.str.contains()` de python y los métodos usados en el ejercicio anterior ademas de `.fillna()`, `.head()`, `.drop()`, `.isnull()`, `.unstack()` y `.sort_values()`, de pandas. Recuerde que para el filtrado en pandas `~ es not`, `& es and` y `| es or` y encerrar entre paréntesis las operaciones lógicas.**</font> 

```
# Consultas en DF

get_connection()
df_content = pd.read_sql('SELECT * FROM content', conn)

df_cantidad = df_content[~df_content['country'].isnull()][['country','type']]
df_cantidad = df_cantidad[~df_cantidad['country'].str.contains(',')]

df_group = df_cantidad.groupby(['country','type'])['type'].count().unstack().fillna(0)
df_group['Total'] = df_group['Movie'] + df_group['TV Show']
df_group = df_group.sort_values('Total',ascending=False).head(10)
df_group.drop('Total', axis= 'columns',inplace=True)

colors = ["#663399","#FF69B4"]  # Crea una paleta de colores personalizada en formato hexadecimal
n = len(df_group.index)
x = np.arange(n)
width = 0.25
value_mul = 0.5
fig, ax = plt.subplots()

for index_value, color in zip(list(df_group.columns), colors):
    p = ax.bar(x - value_mul * width, df_group[index_value], width=width, label=index_value, color=color)
    value_mul = -0.5
    ax.bar_label(p, padding=5, rotation=90)

plt.xticks(x, df_group.index, rotation=80)
ax.set_title('10 países con mayor cantidad de series y películas',  pad=20, fontsize=12)
plt.xlabel('Países', fontsize=9)
plt.ylabel('Cantidad de títulos', fontsize=9)
ax.legend(loc='best')
ax.set_ylim(0, 2300)
plt.show()

close_connection()
```
